```python
# --- Feature Usage Analysis (Meta-style) ---

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate user activity data
np.random.seed(42)
users = [f"user_{i}" for i in range(1, 501)]
features = ["Posts", "Likes", "Messages", "Stories", "Comments"]

data = []
for user in users:
    for feature in features:
        usage_count = np.random.poisson(lam=np.random.randint(5, 50))
        data.append([user, feature, usage_count])

df = pd.DataFrame(data, columns=["user_id", "feature", "usage_count"])

# Total usage per feature
feature_usage = df.groupby("feature")["usage_count"].sum().sort_values(ascending=False)

# Average engagement per user
user_engagement = df.groupby("user_id")["usage_count"].sum().mean()

print("üìä Average engagement per user:", round(user_engagement, 2))
print("\nüîù Feature usage:\n", feature_usage)

# Visualization
plt.figure(figsize=(8,5))
feature_usage.plot(kind="bar", color="skyblue")
plt.title("Total Feature Usage (Simulated Social App)")
plt.xlabel("Feature")
plt.ylabel("Total Interactions")
plt.show()
```
